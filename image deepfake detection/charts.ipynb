```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data for EfficientNetB0
epochs_effnet = [1, 2, 3, 4, 5, 6, 7]
train_acc_effnet = [0.7397, 0.8041, 0.8199, 0.8269, 0.8325, 0.8438, 0.8523]
val_acc_effnet = [0.7975, 0.8070, 0.7803, 0.8018, 0.8123, 0.7673, 0.7968]

# Data for VGG16
epochs_vgg = [1, 2, 3, 4, 5, 6]
train_acc_vgg = [0.7466, 0.8414, 0.8825, 0.9122, 0.9316, 0.9466]
val_acc_vgg = [0.8413, 0.8152, 0.8528, 0.8462, 0.8290, 0.8545]

# Data for New Model
epochs_new = list(range(1, 15))
train_acc_new = [0.6331, 0.859, 0.9003, 0.9194, 0.9303, 0.9371, 0.9435, 0.9491, 0.952, 0.9542, 0.9596, 0.9603, 0.9614, 0.964]
val_acc_new = [0.6125, 0.893, 0.908, 0.9373, 0.923, 0.9243, 0.92, 0.9423, 0.9423, 0.9385, 0.9523, 0.9498, 0.958, 0.9365]





```

```python
# Plot Accuracy Comparison
plt.figure(figsize=(12, 6))
plt.plot(epochs_effnet, train_acc_effnet, 'bo-', label="EfficientNetB0 Train Accuracy")
plt.plot(epochs_effnet, val_acc_effnet, 'b--', label="EfficientNetB0 Val Accuracy")
plt.plot(epochs_vgg, train_acc_vgg, 'ro-', label="VGG16 Train Accuracy")
plt.plot(epochs_vgg, val_acc_vgg, 'r--', label="VGG16 Val Accuracy")
plt.plot(epochs_new, train_acc_new, 'go-', label="CNNv2 Train Accuracy")
plt.plot(epochs_new, val_acc_new, 'g--', label="CNNv2 Val Accuracy")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy Comparison')
plt.legend()
plt.grid()
plt.show()
```

```python
# Bar Chart for Final Accuracy Comparison
models = ['EfficientNetB0', 'VGG16', 'CNNv2']
final_train_acc = [train_acc_effnet[-1], train_acc_vgg[-1], train_acc_new[-1]]
final_val_acc = [val_acc_effnet[-1], val_acc_vgg[-1], val_acc_new[-1]]

x = np.arange(len(models))
width = 0.3

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, final_train_acc, width, label='Train Accuracy', color='blue')
plt.bar(x + width/2, final_val_acc, width, label='Val Accuracy', color='red')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Final Training & Validation Accuracy per Model')
plt.xticks(x, models)
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# Line Chart for CNN Model Performance
plt.figure(figsize=(10, 5))
plt.plot(epochs_new, train_acc_new, 'g-', label='Train Accuracy')
plt.plot(epochs_new, val_acc_new, 'g--', label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('New Model Training & Validation Accuracy')
plt.legend()
plt.grid()
plt.show()

```

```python
# Create a Table Summary
data = {
    "Model": ["EfficientNetB0"] * 7 + ["VGG16"] * 6 + ["CNN Model"] * 14,
    "Epoch": epochs_effnet + epochs_vgg + epochs_new,
    "Train Accuracy": train_acc_effnet + train_acc_vgg + train_acc_new,
    "Validation Accuracy": val_acc_effnet + val_acc_vgg + val_acc_new
}

df = pd.DataFrame(data)
print(df)
```
